```python
import pandas as pd
data = pd.read_csv("C:/Users/HHemery_stg/Desktop/Prog/Datasc/house-prices-advanced-regression-techniques/train.csv")

df = pd.DataFrame(data) 

pd.set_option("display.max_columns",None)

df.head()
```

```python
df.info()
```

```python
print("connaitre la présence des NaN :\n" , df.isna().sum())
```

```python
numeric_cols = df.select_dtypes(include=['number'])

print("médianes des colonnes nombres :\n", numeric_cols.median())
```

```python
import matplotlib.pyplot as plt

plt.hist(df["SalePrice"])
plt.show()
```

```python
numeric_cols.fillna(numeric_cols.median(), inplace=True)

print("nouvelles colonnes numériques : \n", numeric_cols)
```

```python
numeric_cols.hist(bins=30, figsize=(10, 8))
plt.suptitle('Distribution des colonnes numériques')
plt.show()
```

```python
print("Vérificatipn des NaN dans les colonnes numériques : \n" , numeric_cols.isna().sum())
```

```python
categorical_cols = df.select_dtypes(include=['object'])

print("colonnes catégoriques : \n" , categorical_cols)
```

```python
print("NaN Catégoriques : \n", categorical_cols.isna().sum())
```

```python
print("Vérification des modes de la colonne catégorique : \n", categorical_cols.mode())
```

```python
mode_categorical = categorical_cols.mode().iloc[0]

print("mode pour chaque colonnes :\n", mode_categorical )
```

```python
categorical_cols.fillna(mode_categorical, inplace=True)

print("Vérification des remplacements des NaN\n", categorical_cols.isna().sum())
```

```python
import seaborn as sns

for i in categorical_cols:
    unique_values = categorical_cols[i].unique()
    n_colors = len(unique_values)
    palette = sns.color_palette("husl", n_colors=n_colors)
    sns.countplot(x=i, data=categorical_cols, palette=palette, legend=False, hue=i)
    plt.title('Distribution des catégories avant encodage')
    plt.show()
```

```python
####XGboost avec les données numériques

import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

y=numeric_cols.SalePrice

X = numeric_cols.drop(['SalePrice', 'Id'], axis=1)

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=0)

xgb_model = xgb.XGBRegressor(
    objective='reg:squarederror',  # Pour la régression
    n_estimators=100,              # Nombre d'arbres (ou d'étapes de boosting)
    learning_rate=0.1,             # Taux d'apprentissage
    max_depth=3,                   # Profondeur maximum des arbres
    random_state=0
)


xgb_model.fit(Xtrain, ytrain)

y_pred = xgb_model.predict(Xtest)

r2 = r2_score(ytest, y_pred)
print("Coefficient de détermination (R²):\n", r2)
```

```python
mae = mean_absolute_error(ytest, y_pred)
print("Erreur absolue moyenne (MAE):\n", mae)
```

```python
print(y_pred)
```

```python
plt.hist(y_pred)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
palette = sns.color_palette("husl", n_colors=10)
xgb.plot_importance(xgb_model, importance_type='weight', title='Importance des caractéristiques', xlabel='Poids', ylabel='Caractéristiques', max_num_features=10,color=palette)
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x=ytest, y=y_pred)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')  # Ligne de référence
plt.xlabel('Valeurs réelles')
plt.ylabel('Prédictions')
plt.title('Prédictions vs Valeurs réelles')
plt.show()
```

```python
###### XGBoost avec les données numériques et Catégoriques

##Encodage

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()

# Appliquer l'encodage
# Cela produit une matrice sparse qu'il faudra convertir en tableau dense
categorie_col_enc = encoder.fit_transform(categorical_cols).toarray()

# Obtenir les noms des colonnes encodées
encoded_columns = encoder.get_feature_names_out(categorical_cols.columns)

# Convertir en DataFrame
df_col_enc = pd.DataFrame(categorie_col_enc, columns=encoded_columns)

# Concaténer avec tes colonnes numériques dans X
df_final = pd.concat([X.reset_index(drop=True), df_col_enc], axis=1)

print(df_final.head())
```

```python
X = df_final

y = numeric_cols.SalePrice

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=0)

xgb_model = xgb.XGBRegressor(
    objective='reg:squarederror',  # Pour la régression
    n_estimators=100,              # Nombre d'arbres (ou d'étapes de boosting)
    learning_rate=0.1,             # Taux d'apprentissage
    max_depth=3,                   # Profondeur maximum des arbres
    random_state=0
)


xgb_model.fit(Xtrain, ytrain)

y_pred = xgb_model.predict(Xtest)

print(y_pred)
```

```python
r2 = r2_score(ytest, y_pred)
print("Coefficient de détermination (R²):\n", r2)
```

```python
mae = mean_absolute_error(ytest, y_pred)

print("Erreur absolue moyenne (MAE):\n", mae)
```

```python
plt.hist(y_pred)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
palette = sns.color_palette("husl", n_colors=10)
xgb.plot_importance(xgb_model, importance_type='weight', title='Importance des caractéristiques', xlabel='Poids', ylabel='Caractéristiques', max_num_features=10, color=palette)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=ytest, y=y_pred)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')  # Ligne de référence
plt.xlabel('Valeurs réelles')
plt.ylabel('Prédictions')
plt.title('Prédictions vs Valeurs réelles')
plt.show()
```

```python
####Random Forest données numériques

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score

X = numeric_cols.drop(['SalePrice', 'Id'], axis=1)

y=numeric_cols.SalePrice

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=0)

rf = RandomForestRegressor(random_state=0)

# Entraîner le modèle
rf.fit(Xtrain, ytrain)

# Faire des prédictions
y_pred = rf.predict(Xtest)

print(y_pred)
```

```python
print("Score :\n" ,r2_score(ytest, y_pred))
```

```python
print("MEA : \n" , mean_absolute_error(ytest, y_pred))
```

```python
import numpy as np

importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]

# Créer un graphique en barres
plt.bar(range(X.shape[1]), importances[indices],
       color="r", align="center")
plt.xticks(range(X.shape[1]), X.columns[indices], rotation=90)
plt.show()
```
